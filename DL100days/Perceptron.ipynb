```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(
    n_samples=100,
    n_features=2,
    n_informative=2,      
    n_redundant=0,
    n_classes=2,
    random_state=9
)
```

```python
X.shape, y.shape    
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
def perceptron(X, y):
    w1 = w2 = b = 1
    lr = 0.1
    for _ in range(1000):
        for i in range(X.shape[0]):
            z = w1 * X[i][0] + w2 * X[i][1] + b
            if z*y[i] < 0:
                w1 = w1 + lr * y[i] * X[i][0]
                w2 = w2 + lr * y[i] * X[i][1]
                b = b + b * y[i]
    return w1, w2, b
```

```python
w1, w2, b = perceptron(X, y)
```

```python
w1, w2, b
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')

x1 = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
x2 = -(w1 * x1 + b) / w2  # Derived from w1*x1 + w2*x2 + b = 0
plt.plot(x1, x2, 'k--', label='Decision Boundary')

plt.legend()
plt.show()
```
